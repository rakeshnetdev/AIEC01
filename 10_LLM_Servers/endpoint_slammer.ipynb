```python
from langchain_fireworks import ChatFireworks
import nest_asyncio
nest_asyncio.apply()

# REPLACE WITH YOUR OWN ENDPOINT IDENTIFIER
model_endpoint = "accounts/raki-eng-wiz6-tzdwnf/deployments/gvtmt0ab"

llm = ChatFireworks(model=model_endpoint)

response = llm.invoke("How much wood could a wood chuck chuck if a wood chuck could chuck wood?")
print(response.content)
```

Now, let's SLAM IT.

```python
import asyncio

async def send_request(llm, idx):
    try:
        response = await llm.ainvoke(
            f"How much wood could a wood chuck chuck if a wood chuck could chuck wood? (Request {idx})"
        )
        print(f"Response {idx}: {response.content[:60]}...")
    except Exception as e:
        print(f"Request {idx} failed: {e}")

async def main():
    tasks = [send_request(llm, i) for i in range(1, 25)]
    await asyncio.gather(*tasks)

await main()
```

```python
from langchain_fireworks import ChatFireworks

# Use a public Serverless Endpoint (no scale-to-zero waiting required!)
serverless_model = "accounts/fireworks/models/gpt-oss-20b"

# Or if you want to test Llama 3 instead, you can uncomment this:
# serverless_model = "accounts/fireworks/models/llama-v3p1-8b-instruct"

llm_serverless = ChatFireworks(model=serverless_model)

print(f"Testing Serverless Model: {serverless_model}")
response = llm_serverless.invoke(
    "How much wood could a wood chuck chuck if a wood chuck could_chuck wood?"
)
print("\nResponse:")
print(response.content)
```

```python
import asyncio

async def main_serverless():
    print("Slamming the Serverless endpoint with 100 requests...")
    # We pass llm_serverless instead of the dedicated llm
    tasks = [send_request(llm_serverless, i) for i in range(1, 25)]
    await asyncio.gather(*tasks)

await main_serverless()
```
